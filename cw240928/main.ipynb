```python
def chastota(s: str) -> int:
    diction: dict[int: str] = {}

    for chast in s:
        if chast.isalpha():
            if chast in diction:
                diction[chast] += 1
            else:
                diction[chast] = 1

    sortirovano = sorted(diction.items(), key=lambda item: item[1], reverse=True)

    most_common_letter = sortirovano[0][0]
    key = ord(most_common_letter) - ord('о')
    return key

s = """Ш бегдциы, беиежцх еиъывхвцзт иевтбе дычевтояг бежяъежег еи бцчядыиц Явтя Явтянц, ёезвсоцвезт здцнцвц иенде шежнцдты мыёдеа зечцбя, ёеиег зийб зёжсщдйшоял еибйъц-ие дещ. Уие Юцлцж зёжсщдйв з выэцдбя, дц беиежеа ечсбдешыдде ёжешеъяв шжыгх, зяъх ёещжйэыддса ш ъжыгеий. Ш бегдций шеоыв ёеэявеа нывешыб, ш зыжег зфжийбы, з ёжежылеф ёеъ гсобеа, еибйъц иежнцв бвенеб жйчцобя, ш зыжег эы эявыиы, з гыъдсгя ёйщешямцгя, з щевсг, бцб бевыде, ныжыёег я з дыечрхиде ояжебягя я щйзисгя жйзсгя з ёжезыътф чцбыдчцжъцгя, яю беиежсл бцэъеа зицве чс дц ижя чежеъс. Юцлцж ды зицжцвзх яюгыдяит ды иевтбе ъцддеще ыгй чещег ечжцюц, де я зшеыще безифгц, ш беиежег леъяв ш ъыжышды. Ёвциты ыгй оявезт ёе шсшыюыддегй яг яю ъыжышдя ечжцюмй. Зыжса зфжийб я эявыи джцшявязт ыгй я ёеиегй, ние ш уиеа ёевйкежгыддеа еъыэъы ед шяъыв звцчеы шезёегядцдяы вяшжыя, беиежйф ед дезяв дыбещъц, ёжешеэцх ёебеадсл щезёеъ ш мыжбешт явя ш щезия; ц вяшжых ш шезёегядцдяхл ыще чсвц ыъядзишыддеф ёжыъзицшяиывтдямыф ъезиеядзишц ъегц Ечвегешсл. Чевыы дяние ды дцёегядцве зицжябй чцжзбеще ояжебеще я ёебеадеще чсиц ш щвйоя ъыжышдя. Зицжсы щезёеъц йгыжвя, кцгявтдсы ёежижыис езицвязт ъегц я, нца, шцвхфизх щъы-дячйът дц ныжъцбы; ёжыъцдях е зицжяддег чсиы я шцэдезия кцгявяя шзь щвелдйи явя эяшйи иевтбе ш ёцгхия дыгдещял, езицшоялзх ш ъыжышды эы зицжябеш. Ёеуиегй ъвх Юцлцжц ъежещ чсв зыжса зфжийб: ш дыг ъц ыпы ш беы-бцбял ёжяюдцбцл, зелжцдяшоялзх ш вямы я гцдыжцл чцжядц, дцёегядцшоял ыще жеъяиывыа, я ш ыще бцёжяюцл, дц беиежсы леих ед я шежнцв, я ёже зычх я шзвйл, де беиежсы гыэъй иыг йшцэцв шдйижыдды, бцб ёжехшвыдяы чцжзбеа шевя, щезёеъзбеще ёжцшц, шяъыв ед звцчсы дцгыбя дц еиэяшоыы шывяняы.""".lower()

print('Павел Георгиевич, я вам отвечаю это правильный ответ ->', chastota(s))
```
