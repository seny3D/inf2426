Омега - всевозможные упорядоченные тройки чисел от 1 до 6 ( (1,1,1), (1,1,2), (1,1,3), ... (6,6,5), (6,6,6) ). Таких будет 6^3 написать название формулы.

Сигмаалгебра - Множество всех подмножеств Омеги. Например: A = {(1,1,1)}, B = {(1,1,1), (1,1,2),  (1,1,3), (1,1,4), (1,1,5), (1,1,6)} и тд

Вероятностное пространство: элементы Омега ( (1,1,1), (1,1,2), (1,1,3), ... (6,6,5), (6,6,6) ) имеют одинаковую вероятность, считаем их элементарными событиями. А занчит веростность P((1,1,1)) = 1/6^3 
Для любого события из сигма алгебры его вероятность будет считать  по формуле: P(A) = |A|/6^3
