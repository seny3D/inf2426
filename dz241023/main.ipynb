```python
# Плотность графика равномерного распределения
import numpy as npy
import matplotlib.pyplot as plot
from typing import Any

a: int = 0
b: int = 1

rand: Any = npy.random.uniform(a, b, 1000)

plot.hist(rand, bins=40, density=True)

x: Any = npy.linspace(a, b, 100)
plot.plot(x, npy.ones_like(x) / (b - a))
plot.grid()
plot.show()
```

```python
# Плотность графика показательного распределения
import numpy as npy
import matplotlib.pyplot as plot
from typing import Any

lambd: int = 1

rand: Any = npy.random.exponential(1 / lambd, 1000)

plot.hist(rand, bins=40, density=True)

x: Any = npy.linspace(0, npy.max(rand), 100)
plot.plot(x, lambd * npy.exp(-lambd * x))
plot.grid()
plot.show()
```

```python
# График равномерного распределения
import numpy as npy
import matplotlib.pyplot as plot
from typing import Any

x: Any = npy.linspace(0, 2, 100)
y: Any = npy.clip(x, 0, 1)

plot.plot(x, y)
plot.grid()
plot.show()
```

```python
# График показательного распределения
import numpy as npy
import matplotlib.pyplot as plot
from typing import Any

x: Any = npy.linspace(0, 5, 100)
y: Any = 1 - npy.exp(-x)

plot.plot(x, y)
plot.grid()
plot.show()
```

```python
# График распределения Бернулли
import numpy as npy
import matplotlib.pyplot as plot
from typing import Any

a: float = 0.5

x: Any = npy.array([0, 1, 2])

func: Any = npy.array([1 - a, 1, 1])

plot.step(x, func, where='post')
plot.show()
```
