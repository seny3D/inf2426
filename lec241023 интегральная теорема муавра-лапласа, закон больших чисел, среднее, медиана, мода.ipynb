# Индикатор множества

$$
I_{A}(x) =
\left\{ 
    \begin{array}{l}
        1\ если\ x \in A, &\\
        0\ если\ x \notin A
    \end{array} 
\right.
$$

# Интегральная теорема Муавра-Лапласа

$$
P\Big\{a \leq \frac{\mu - np}{\sqrt{npq}} < b \Big\} \sim \frac{1}{\sqrt{2\pi}} \int_{a}^{b} e^{-\frac{x^2}{2}}dx
$$

эта теорема позволяет решать среди прочего следующие задачи

##  I
Чему равна вероятность того, что частота наступления события $A$ отклонится от вероятности $p$ не более чем на $\alpha$

$$
P\Big\{ \Big|\frac{\mu}{n} - p\Big| \le \alpha \Big\}
$$

## II
Какое наименьше число испытаний нужно произвести для того, чтобы с вероятностью, не меньше $\beta$, частота отклонялась о вероятности не более чем на $\alpha$. Нужно определить $n$ из неравенства

$$
P\Big\{ \Big|\frac{\mu}{n} - p\Big| \le \alpha \Big\} \le \beta
$$

###  III
При данной вероятности $\beta$ и числе испытаний $n$ требуется определить границу возможных изменений $\big|\frac{\mu}{n}-p\big|$. Иными словами зная $\beta$ и $n$, нужно найти $\alpha$, для которого
$$
P\Big\{ \Big|\frac{\mu}{n} - p\Big| \le \alpha \Big\} = \beta
$$

# Функция распределения случайной величины

$$
F_\mu(x) = P(\mu \leq x)
$$

## Свойства
$$
P(a < \mu \leq b) = P(\mu \leq b) - P(\mu \leq a) = F_\mu(b) - F_\mu(a)
$$

## Примеры распределений
### Равномерное распределение
$$
F_\mu(x) =
\left\{ 
    \begin{array}{l}
        0\ при\ x \le 0, &\\
        x\ при\ 0 < x < 1, &\\
        1\ при\ x \ge 1.
    \end{array} 
\right.
$$

### Показательное распределение

$$
F_\mu(x) =
\left\{ 
    \begin{array}{l}
        0\ при\ x \le 0, &\\
        1 - e^{-\lambda x}\ при\ x > 0
    \end{array} 
\right.
$$
где $\lambda$ это параметр распределения

### Распределение Бернулли
$$
F_\mu(x) =
\left\{ 
    \begin{array}{l}
        0\ при\ x < 0, &\\
        1-p\ при\ 0 \le x < 1, &\\
        1\ при\ x \ge 1.
    \end{array} 
\right.
$$
где $p = P(\mu = 1)$ 

# Плотность

Если существует такая функция $p_\mu(x) \ge 0$, что для произвольных $a < b$

$$
P(a \le \mu \le b) = \int_a^b p_\mu(x)dx
$$
то говорят что случайная величина $\mu$ имеет плотность $p_\mu(x)$

## Свойства
Если плотность существует, то
$$
p_\mu(x) = F'_\mu(x)
$$

плотности не существует у дискретных случайных величин

## Примеры плотности
### Плотность равномерной случайной величины
$$
p_\mu(x) = I_{[0,1]}
$$

### Плотность показательной случайной величины
$$
p_\mu(x) = \lambda e^{-\lambda x} I_{[0, +\infty)}
$$


# Мат. ожидание

## Для дискретных случайных величин
мат. ожидание это взвешенная по вероятности сумма значений случайной величины

$$
M_\mu = \sum_{i} \mu_i p_i
$$

## Для непрерывных случаных величин

$$
M_\mu = \int_{-\infty}^{+\infty} xp_\mu(x) dx
$$
