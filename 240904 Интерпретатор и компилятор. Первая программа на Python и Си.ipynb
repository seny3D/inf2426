# Для Python 

1. Скачиваем и устанавливаем интерпретатор https://www.python.org/downloads/
2. Выбираем текстовый редактор
3. Настраиваем текстовый редактор, чтобы вместо TAB было четыре пробела
4. Настраиваем текстовый редактор на работу в кодировке utf-8
5. Создаем через текстовый редактор нужный файл с расширением .py, например first_python_program.py
6. Записываем в файл программу на языке python

```python
# Это комментарий.
# Все что правее решетки, включая саму решетку, 
# игнорируется интерпретатором


# Определение функции 
# def имя_функции (аргументы) -> возвращаемое значение
def add_two() -> None:
    # То что находится внутри функции должно быть смещено на 4 пробела
    adder: int = 2  # создаем переменную типа int

    # считываем данные с консоли (стандартного ввод)
    number_in: int = int(input())
    
    # создаем переменную и кладем в нее результат сложения
    result: int = number_in + adder

    # вывод строки на консоль (стандартный вывод)
    # используем f-строку для удобного форматирования
    # number_in и result преобразуются сначала в строки потом вставляются
    # в f-строку
    print(f"Вы ввели число {number_in}, на 2 больше это {result}")
    return None  # можно не писать, по умолчанию и так возращается None

# Встроенная переменная __name__ имеет значение либо '__main__',
# если этот файл является тем, кто передан интерпретатору на выполнение,
# либо значение __name__ равно имени файла, если он подключается как
# модуль к другим файлам.
# Мы зайдем внутрь условия, только если этот файл является
# точкой входа в программу.
if __name__ == '__main__':
    # вызывается функция add_two
    add_two()

```

7. Запускаем программу на выполнение передавая ее интерпретатору в качестве аргумента

```shell
python first_python_program.py
```

# Для Си 

1. Скачиваем и устанавливаем компилятор https://www.codeblocks.org/downloads/binaries/
2. Выбираем текстовый редактор
3. Создаем через текстовый редактор нужный файл с расширением .с, например first_с_program.с
4. Записываем в файл программу на языке си.

```c
// Все что правее двойной косой черны игнорируется
// вклюая ее саму.
// Это называется однострочный комментарий.

/*
 * Все что внутри косой черты и звездочки игнорируется.
 * Это называется многострочный комментарий.
 * Звездочки вначале не обязательны, но с ними удобнее понимать, 
 * что это часть является комментарием.
 */
 
// подключение заголовочного файла сторонних функций
// мы из него будем использовать функции scanf и printf
#include <stdio.h>

// Объявление функции: тип_возвращаемого_значения имя_функции(параметры)
// Притом в программе которую
int main()
{  // Начало тела функции
    int adder = 2;  // Объявление переменной типа int и присваивание значения 2
    int number_in, result;
    
    // Вызов функции scanf, она ожидает в качестве аргументов формат ввода и 
    // адреса переменных.
    // Считывание целого числа с консоли и запись по адресу переменной number_in
    scanf("%d",  &number_in);
    // вычисление результата сложения и сохранение в переменную
    result = number_in + adder;
    
    // Вызов функции printf, она ожидает строки форатирования и 
    // значения переменных
    // Вывод на консоль строки
    printf("Вы ввели число %d, на 2 больше это %d\n", number_in, result);
    return 0;
}  // Конец тела функции
```

5. Компилируем программу и получаем exe файл a.out

```shell
gcc first_c_program.c
```

6. Запускаем экзешник

```shell
./a.out
```
